```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P = 80 # force exerted by actuator (kgf)
E = 200 # Young' modulus (GPa)

c = np.arange(0,50,0.1)
L = np.pi *(c**3)/(P*9.8)*E*0.002
```

```python
plt.plot(c, L)
plt.xlabel('Diamter of Axis')
plt.ylabel('Length of Axis')
plt.show()
```

```python
plt.plot(L, c)
plt.ylabel('Diamter of Axis')
plt.xlabel('Length of Axis')
plt.show()
```
